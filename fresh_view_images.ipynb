```python
from scipy.io import loadmat
import matplotlib.pyplot as plt

# Load the .mat file
data = loadmat('Project datasets/Matlab files/Images/Background subtracted/Cell_01_In_532_BG.mat')

# Inspect the loaded data
print(data.keys())  # This will list all variables inside the .mat file

# Extract the image data
# Assuming your image data is stored under a known variable name, e.g., 'image_data'
image_data = data['image_data']

# Display the image (if it's a 2D or 3D array compatible with an image format)
plt.imshow(image_data, cmap='gray')  # You can change 'gray' to another color map if it's a color image
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Step 1: Load the .mat file
file_path = 'Project datasets/Matlab files/Images/Background subtracted/Cell_01_In_532_BG.mat'  # Update with your file path
mat_data = loadmat(file_path)
```

```python
type(mat_data)
```

```python
mat_data.keys()
```

```python
# Step 2: Access the MATLAB object
spec_data_object = mat_data['Large_Area_Scan_003_SpecData_1_BR_BR_BR_Sub_BG']

# Step 3: Convert the object to a NumPy array and extract nested data
spec_data_array = np.array(spec_data_object)
nested_data = spec_data_array[0, 0]
```

```python
type(nested_data)
```

```python
# Step 4: Extract image data and image size
image_data = nested_data['data']  # Image pixel values
image_size = tuple(nested_data['imagesize'][0])  # Image dimensions as (height, width)
```

```python
# Step 5: Reshape the data to match the image size
if image_data.shape == image_size:
    image_data = image_data.reshape(image_size)
```

```python
# Step 6: Save the image data as a NumPy array for direct access in Python
output_array_path = 'extracted_image.npy'
np.save(output_array_path, image_data)
print(f"Image data saved as NumPy array to: {output_array_path}")
```

```python
import numpy as np
from scipy.io import loadmat
import plotly.express as px

# Step 1: Load the .mat file
file_path = 'Project datasets/Matlab files/Images/Background subtracted/Cell_01_In_532_BG.mat'
mat_data = loadmat(file_path)

# Step 2: Access the MATLAB object
spec_data_object = mat_data['Large_Area_Scan_003_SpecData_1_BR_BR_BR_Sub_BG']

# Step 3: Convert the object to a NumPy array and extract nested data
spec_data_array = np.array(spec_data_object)
nested_data = spec_data_array[0, 0]

# Step 4: Extract image data and image size
image_data = nested_data['data']  # Image pixel values
image_size = tuple(nested_data['imagesize'][0])  # Image dimensions as (height, width)

# Step 5: Reshape the data to match the image size
if image_data.shape == image_size:
    image_data = image_data.reshape(image_size)

# Step 6: Visualize with Matplotlib
plt.imshow(image_data, cmap='gray')
plt.axis('off')
plt.title("Visualized Image from .mat File")
plt.show()
```
