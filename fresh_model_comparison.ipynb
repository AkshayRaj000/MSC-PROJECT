```python
import pandas as pd

# Define the model performance data
data = {
    "Model": ["DAE", "RCAN", "1D CNN", "1D U-Net", "Wavelet Autoencoder"],
    "RMSE": [0.0028103442014676024, 0.010542593305582507, 0.00773991020268567, 0.01026457756696739, 0.0038209989456255322],
    "MAE": [0.0021861330413020202, 0.00916892443738763, 0.0063037419849859035, 0.009015051408874914, 0.002935405333472683],    
    "SNR (dB)": [25.88, 11.5191950348385, 14.203425660220105, 11.7513224835455, 20.334605745066252], 
    "PSNR (dB)": [26.592247040661636, 18.773648631440842, 21.45787925682245, 19.005776080147843, 27.589059341668595], 
    "SSIM": [0.996179331957553, 0.9462058048537656, 0.9693361549421945, 0.9458616857856254, 0.9941520296543565]    
}

# Create a DataFrame
performance_df = pd.DataFrame(data)

# Display the table
performance_df
```

```python
import matplotlib.pyplot as plt

# Extract the model names
models = performance_df["Model"]

# Plotting each metric as a separate bar plot
metrics = ["RMSE", "MAE", "SNR (dB)", "PSNR (dB)", "SSIM"]

# Set up subplots for each metric
for metric in metrics:
    plt.figure(figsize=(8, 5))
    plt.bar(models, performance_df[metric], color='skyblue')
    plt.xlabel("Models")
    plt.ylabel(metric)
    plt.title(f"Comparison of {metric} Across Models")
    plt.xticks(rotation=45)
    plt.show()
```


Based on our comparison, the Denoising Autoencoder (DAE) was selected as the best model for spectral denoising. It achieved the lowest RMSE (0.0028) and MAE (0.0022), indicating minimal error in noise reduction. Additionally, the DAE had the highest SNR (25.88 dB) and PSNR (26.59 dB), demonstrating its strong capability to enhance signal clarity. With an SSIM of 0.996, it also outperformed other models in preserving structural integrity, making it the most balanced and effective choice for this denoising task.
